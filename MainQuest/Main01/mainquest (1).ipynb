```python
import random

class Account:
    account_count = 0
    
    def __init__(self,owner,balance):
        self.bank = 'sc은행'
        self.owner = owner
        self.balance = balance
        self.account_number = self.generate_account_number()
        self.deposit_count = 0
        self.deposit_history_list = []
        self.withdraw_history_list = []
        Account.account_count += 1
        
        
    def generate_account_number(self):
        first = str(random.randint(100, 999))
        second = str(random.randint(10, 99))
        third = str(random.randint(100000, 999999))
        return f"{first}-{second}-{third}"
    
    @classmethod
    def get_account_num(cls):
        return cls.account_count
    
    def deposit(self,amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1
            self.deposit_history_list.append(amount)
            if self.deposit_count % 5 == 0:  
                interest = self.balance * 0.01
                self.balance += interest
                self.deposit_history_list.append(('이자', interest, self.balance))
        else:
            print("입금은 최소 1원 이상만 가능합니다.")
            
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            self.withdraw_history_list.append(amount)
        else:
            print("계좌의 잔고 이상으로 출금할 수는 없습니다.")
    
    def display_info(self):
        balance_str =f"{self.balance:,}"
        print(f"은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {balance_str}원")
        
    def deposit_history(self):
        print(f"입금 내역: {self.deposit_history_list}")
        
    def withdraw_history(self):
        print(f"출금 내역: {self.withdraw_history_list}")
```

```python
accounts = [ Account("고객1", 1000),
    Account("고객2", 30000000),
    Account("고객3", 50000000)]

for account in accounts:
    if account.balance >= 1000000:
        account.display_info()
```

```python
my_account = Account('고객1', 1000)

for i in range(10):
    my_account.deposit(1000)
my_account.deposit(-1000)
my_account.withdraw(300000)
my_account.withdraw(3000)

my_account.deposit_history()
print('-'* 50)
my_account.withdraw_history()
```

```python
class character:
     def __init__(self, name, level, health, attack_power, defense):
        self.name = name
        self.level = level
        self.health = health
        self.attack_power = attack_power
        self.defense = defense
        
     def is_alive(self):
        return self.health > 0
    
    #데미지를 받아 쳬력을 감소시키는 메서드
    
     def take_damage(self, damage):
        actual_damage = damage - self.defense
        if actual_damage > 0:
            self.health -= actual_damage
        print(f"{self.name}의 체력: {self.health}")
        
        #타겟에 데미지
     def attack_target(self, target):
        damage = random.randint(1, self.attack_power)
        print(f"{self.name}이(가) {target.name}에게 {damage} 만큼 공격했다...!")
        target.take_damage(damage)
        
    #클래스를 상속받는 plyaer메서드
    
class Player(character):
     def __init__(self, name):
        super().__init__(name, level=1, health=100, attack_power=25, defense=5)
        self.experience = 0

        
     def gain_experience(self, exp):
        self.experience += exp
        print(f"{self.name}이(가) 경험치 {exp} 획득!")
        self.level_up()
    
     def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack_power += 10
            self.defense += 5
            self.health = 100  # Reset health on level up
            self.experience = 0  # Reset experience
            print(f"{self.name}이(가) 레벨업 했다! 현재 레벨: {self.level}, 공격력: {self.attack_power}, 방어력: {self.defense}")
            
#몬스터 클래스
class Monster(character):
     def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack_power = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, health, attack_power, defense)
        
#배틀함수
def battle(player, monster):
    print(f"\n{monster.name} 과의 전투를 시작합니다.")
    
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)
        if not player.is_alive():
            print("플레이어가 사망했습니다. 게임 오버!")
            return False
    
    print("전투 승리!")
    print(f"{monster.name} 을(를) 이겼다!")
    player.gain_experience(monster.level * 20) 
    return True

#메인함수

def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}
    
    player_name = input("플레이어의 이름을 입력하세요: ")
    player = Player(player_name)
    
    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)
        if not battle(player, monster):
            break
if __name__ == "__main__":
    main()
```
